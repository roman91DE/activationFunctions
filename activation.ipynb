```python
from matplotlib import pyplot as plt
import numpy as np
import numpy.typing as npt
from typing import Callable
```

# Activation Functions for Deep Learning

```python
def plot_function(f: Callable) ->None:
    X = np.linspace(-10,10,100)
    Y = f(X)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.plot(X, Y)
```

## Sigmoid Activation

$$f(x) = \frac{1}  {1 + e^{-x}}$$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

plot_function(sigmoid)
```

## Hyperbolic tangent Activation
$$f(x) = tanh(x) = \frac{2}  {1 + e^{-2x}}-1$$

```python
def tanh(x):
    return np.tanh(x)

plot_function(tanh)
```

## ReLU Activation
Rectified linear unit activation

$$
f(x) = \begin{cases}
    0\text{, for } x < 0 \\
    x\text{, otherwise}
    \end{cases}
$$

```python
def relu(x: npt.ArrayLike):
    return np.maximum(0,x)


plot_function(relu)
```
